## Objective

- Illustrate the evolution of the mean temperature and viscosity of the mantle of a terrestrial body with a stagnant lid or mobile lid regime of convection

## Background

- The thermal evolution of the interior of terrestrial bodies is governed by a balance between the heat accumulated in the interior during planetary formation and generated through the decay of radioactive isotopes and the heat that is lost at the surface and radiated to space. Such balance strongly depends on the way heat is transported in the mantle by solid-state convection. While a *mobile lid* regime such as that of the Earth with plate tectonics provides a very efficient mechanism to cool the interior and remove heat, the *stagnant lid* regime, which characterizes the other  terrestrial planets of the solar system, is less efficient. The notebook illustrates this idea making use of a simple model of *parameterized* convection that allows us to simulate the evolution of the mean mantle temperature and viscosity of a given body over the age of the solar system.

```python
# Select planet
planet = plc("Mars", debug=0)
mars = planet
```

```python
# Run thermal evolution 
# Input parameters are: planet (e.g. mars), mode of convection ('sl' for stagnant lid, 'ml' for mobile lid),
# initial mantle temperature (e.g. 1800 K), and reference viscosity (e.g. 1e21 Pa s)
# See thermal_evolution.py for details
time, temp, visc = te.thermal_evolution(mars, 'sl', 1800., 1e21)
```
