# 1. Introduction to Planetary Data

Planetary mission data are all public and copyright-free, after a variable period of time, in which experiment teams have the exclusive right to work with data produced by their instruments and publish scientific papers. Once this embargo period is over, all data are available on archives (once as CDs or DVDs, now exclusively online, as FTP, HTTP and throughout web services/applications).

This activity/assignment will go through the processing/reduction steps of a sample remote sensing instrument, a camera on board an active NASA mission (Mars Reconnaissance Orbiter).

## Map projections in ISIS

### Isis Map template

The following is a test map template for ISIS3, e.g. for the Moon

`  Group = Mapping
    ProjectionName     = Equirectangular
    CenterLongitude    = 0.0
    TargetName         = Moon
    EquatorialRadius   = 1737400.0 <meters>
    PolarRadius        = 1737400.0 <meters>
    LatitudeType       = Planetocentric
    LongitudeDirection = PositiveEast
    LongitudeDomain    = 180
    PixelResolution    = 1 <meters/pixel>
    CenterLatitude     = 0.0 <DEG>
End_Group
`

For Mars, a map template, for an Equirectangular projection, one would use

`
Group = Mapping
  ProjectionName     = Equirectangular
  CenterLongitude    = 0.0
  TargetName         = Mars
  EquatorialRadius   = 3396190.0 <meters>
  PolarRadius        = 3396190.0 <meters>
  LatitudeType       = Planetocentric
  LongitudeDirection = PositiveEast
  LongitudeDomain    = 180
  CenterLatitude     = 0.0 <DEG>
End_Group
`

"PixelResolution " is optional and if not specified, the optimal resolution/cell size is computed by the system itself.

# 3. Workflow with real data
CTX data from NASA MRO are used for the 

## 3.1 Level0 Processing (PDS data import)
`mroctx2isis` imports a CTX EDR (experiment data record, i.e. uncalibrated) image into ISIS3. It needs both an input and an output
(a normal PDS image, regardless the level, would be imported with `pds2isis`)

Example:

`mroctx2isis from=INPUT.IMG to=CTX_lev0.cub`

## 3.1a SPICE (pointing, ancillary info) initialization. 

It needs an input (the relevant info is stored in the ASCII ISIS3 header, and one could check the before/after with a more CTX_lev0.cub.

`spiceinit` initializes all geometry and time-related information (e.g. position and pointing of spacecraft). SPICE stands for:

**S** - Spacecraft ephemeris, given as a function of time

**P** - Planet, satellite, comet, or asteroid ephemerides

**I** - Instrument description kernel

**C** - Pointing kernel

**E** - Events kernel, summarizing mission activities

More info on http://naif.jpl.nasa.gov/naif/

Example:

`spiceinit from=CTX_lev0.cub`


## 3.2 Level1 Processing (radiometric calibration)
Calibration of the radiometry (pixed Digital Number / value) is achieved through calibration tables prvoided by ISIS3

`ctxcal from=CTX_lev0.cub to=CTX_lev1.cub` 

and subsequently

`ctxevenodd from=CTX_lev1.cub to=CTX_lev1eo.cub`

## 3.3 Level2 Processing (Map projection)

Once images have been radiometrically calibrated, they can be geometriclaly calibrated and map projected. The choice of the map projection can be specified with a map_template file (ASCII). If no map_template is provided, the defautl prosection for the image location is choosen, e.g. a Sinusoidal projection centered at the longitude of the center of the image.

**without specyfing** a map_template* one would run:

`cam2map from=CTX_lev1eo.cub to=CTX_lev2.cub`

**specyfing** a map_template would be done with:

`cam2map from=CTX_lev1eo.cub to=CTX_lev2.cub map=map_template.txt`

# 4. Simple mosaicking
Mosaicking **REQUIRES** that **ALL** images have the **same pixel size / resolution**, i.e., either:

a) a map_template contains info on pixel resolution AND the `pixres=map` option in cam2map is choosen
b) a map_template does not contain info on pixel resolution and the `pixres=mpp resolution=50` (or any number) is chosen as option on cam2map


**Creation of list**
`
/bin/ls A.cub > image-list.txt
`

*It is advised to inspect the .txt file with `gedit` and eventually get rid of the file size (e.g. `3M`) in the text, before using it, in case you did not use `/bin/ls` but just `ls`, which is an alias...*

**Creation of mosaic**
`
autmos fromlist=image-list mosaic=MOSAIC.cub
`


# 5. Equalization + mosaicking

Before mosaicking image tone can be matched with `equalizer`

supposed we have A.cub and B.cub which are 2 overlapping images, and one would use A.cub as the reference image to be tone-matched by B.cub.

one should build 2 lists, one of the images to be equalised (clearly A and B for this example, potentially many more in a real case):

**Creation of list**
`
/bin/ls A.cub > image-list.txt
/bin/ls B.cub >> image-list.txt 
`

*It is advised to inspect the .txt file with `gedit` and eventually get rid of the file size (e.g. `3M`) in the text, before using it.*

**Creation of "hold list" as reference**
`
ls A.cub > hold-list.txt
`

**Equalization**
`
equalizer fromlist=image-list.txt holdlist=hold-list.txt
`




# 6. Exporting

## 6a GDAL/ORG + Proj4

GDAL utilities can be used and documentation, syntax and example are on:

http://www.gdal.org/gdal_translate.html

`gdal_translate` 

The option `-ot byte -scale` outputs to 8Bit (255 tones of grayscale)

e.g. 

`gdal_translate INPUT.cub OUTPUT.tif -ot byte -scale `

or:

`gdal_translate INPUT.cub OUTPUT.jpg -ot byte -scale -of JPEG `

## 6b isis2std

One can export ISIS3 .cub files with the ISIS3 program `isis2std`, e.g.:

`isis2std from=MOSAIC.cub to=MOSAIC-export.png format=png stretch=linear minpercent=0.2 maxpercent=99.8 ` 

once exported, data can be viewed as graphic file with:

`
okular MOSAIC-export.png
`

or alike.

# 7. Workflow - summary

`

ctxcal from= to=
ctxevenodd from= to=
reduce from= to= mode=scale lscale=10 sscale=10 #OPTIONAL and the lscale, sscale are arbitrary
cam2map from= to= map= pixres=map
equalizer fromlist= holdlist=
automos fromlist= mosaic=
gdal_translate INPUT.cub OUTPUT.png OR isis2std from= to=
`
