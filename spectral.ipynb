# Spectra basic handling

## Spectra loading
Sample spectra are loaded from ASCII tab- or space-separated files (visible in [input directory](spectral_input/). Spectra are imported with [pandas](http://pandas.pydata.org). 2 of them are lab spectra (respectively montmorillonite and nontronite) and 2 MRO CRISM spectra colleccted (with a 5x5 pixel averaging) on Mars, where occurrence of those mineral is known from [literature](ADD URL !!!)

```python
# Import spectra

import matplotlib
from matplotlib import pyplot as plt
%matplotlib inline

import numpy as np
import pandas as pd


specalp = pd.read_csv('spectral_input/FRTC596_537-220-5x5_Al-phyllo_montm.txt', 
                      skipinitialspace=True, sep=' ',
                      skiprows=3, header=None)

libraryalp = pd.read_csv('spectral_input/montmorillonite-1292F35-RELAB.txt', 
                      skipinitialspace=True, sep=' ',
                      skiprows=3, header=None)

specfep = pd.read_csv('spectral_input/FRTC596_537-220-5x5_Fe-phyllo_nontr.txt', 
                         skipinitialspace=True, sep=' ',
                         skiprows=3, header=None)

libraryfep = pd.read_csv('spectral_input/nontronite-CBJB26-RELAB.txt', 
                         skipinitialspace=True, sep=' ',
                         skiprows=3, header=None)
```

## Spectra plotting

```python
fig = plt.figure(figsize=(10,10))
ax1 = fig.add_subplot(211)
plt.xlim(0.5, 2.6)
plt.ylim(0, 1.6)
plt.title('Montmorillonite')

x1 = specalp[0]
y1 = specalp[1]
plt.plot(x1,y1, '.', label="CRISM spectrum", color='b')

x2 = libraryalp[0]
y2 = libraryalp[1]

plt.plot(x2,y2, '.', label="Laboratory spectrum", color='g')

# plt.legend(loc='upper left')

# Montmorillonite
ax1.vlines(1.41, 0, 1.2, colors='red', linestyles='solid', label='1.41') # 1.41 µm
ax1.vlines(1.9, 0, 1.2, colors='blue', linestyles='solid', label='1.9') # 1.9 µm
ax1.vlines(2.21, 0, 1.2, colors='green', linestyles='solid', label='2.21') # 2.21 µm
plt.legend(loc='upper left')

##############################################
ax2 = fig.add_subplot(212)
plt.xlim(0.5, 2.6)
plt.ylim(0, 1.6)
plt.title('Nontronite')

x3 = specfep[0]
y3 = specfep[1]
plt.plot(x3,y3, '.', label="CRISM spectrum", color='b')

x4 = libraryfep[0]
y4 = libraryfep[1]
plt.plot(x4,y4, '.', label="Laboratory spectrum", color='g')

# Plot vertical lines for key adsorbtion bands
# e.g. from http://authors.library.caltech.edu/34903/2/Bishop.SOM.pdf

# Nontronite
ax2.vlines(1.42, 0, 1.2, colors='red', linestyles='solid', label='1,42') # 1.42 µm
ax2.vlines(1.9, 0, 1.2, colors='blue', linestyles='solid', label='1.9') # 1.9 µm
ax2.vlines(2.29, 0, 1.2, colors='green', linestyles='solid', label='2.29') # 2.29 µm
ax2.vlines(2.41, 0, 1.2, colors='black', linestyles='solid', label='2.41') # 2.41 µm
plt.legend(loc='upper left')






plt.show()
```

## Variation of spectral response at Infrared wavelenghts

Certain minerals are featureless in VNIR, but feature-rich e.g. in TIR, e.g quartz.
2 spectra of the same mineral, collected at different wavelenghts are loaded (all downloaded locally from the USGS Spectroscopy Lab database):

- [Quartz (Visibe and Near Infrared)](http://speclab.cr.usgs.gov/spectral.lib06/ds231/ASCII/M/quartz_hs32.19141.asc)
- [Quartz (Short Wave and Thermal Infrared](http://speclab.cr.usgs.gov/spectral.lib06/ds231/ASCII/M/quartz_gds74.19086.asc)

## Loading spectra of Quartz in VNIR & SWIR/TIR

```python
import matplotlib
from matplotlib import pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd


qtzvnir = pd.read_csv('spectral_input/quartz_hs32.19141.asc', sep=" ",
                      skipinitialspace=True, header=13)

qtztir = pd.read_csv('spectral_input/quartz_gds74.19086.asc', sep=" ",
                      skipinitialspace=True, header=13)

# uncommenting the line below would show the ASCII file as table loaded in Pandas
# qtzvnir
```

# Plots Quartz spectra

```python
# VNIR
fig = plt.figure(figsize=(10,10))
ax1 = fig.add_subplot(211)
plt.xlim(0.2, 2.9)
plt.ylim(0, 1)
plt.title('Quartz - VNIR', fontsize=14)
plt.xlabel("Wavelenght in $\\mu$m", fontsize=10)
plt.ylabel("Reflectance", fontsize=10)

xq1 = qtzvnir['wavelength']
yq1 = qtzvnir['reflectance']

# plot for comparison an earlier spectrum
# x2 = libraryalp[0]
# y2 = libraryalp[1]
# plt.plot(x2,y2, '.', linewidth=2.0, label="Montmorillonite VNIR", color='g')

plt.plot(xq1,yq1, '-', linewidth=2.0, label="Qtz VNIR", color='b')


plt.legend(loc='upper left')

######################

# TIR
ax2 = fig.add_subplot(212)
plt.xlim(5, 24)
plt.ylim(0, 1)
plt.title('Quartz - SWIR / TIR', fontsize=14)
plt.xlabel("Wavelenght in $\\mu$m", fontsize=10)
plt.ylabel("Reflectance", fontsize=10)

xq2= qtztir['wavelength']
yq2 = qtztir['reflectance']
plt.plot(xq2,yq2, '-', linewidth=2.0, label="Qtz TIR", color='r')

plt.show()
```
